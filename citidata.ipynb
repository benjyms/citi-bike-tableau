This workbook will be used to extract, tranform and load the 2020 Citibike Trip and Ride Data for analysis in tableau

```python
# importing all libraies needed
import pandas as pd
import numpy as np
import csv
import matplotlib.pyplot as plt
```

```python
# read in the january data
january_df = pd.read_csv("Data\JC-202001-citibike-tripdata.csv")
february_df = pd.read_csv("Data\JC-202002-citibike-tripdata.csv")
march_df = pd.read_csv("Data\JC-202003-citibike-tripdata.csv")
april_df = pd.read_csv("Data\JC-202004-citibike-tripdata.csv")
may_df = pd.read_csv("Data\JC-202005-citibike-tripdata.csv")
june_df = pd.read_csv("Data\JC-202006-citibike-tripdata.csv")
july_df = pd.read_csv("Data\JC-202007-citibike-tripdata.csv")
august_df = pd.read_csv("Data\JC-202008-citibike-tripdata.csv")
september_df = pd.read_csv("Data\JC-202009-citibike-tripdata.csv")
october_df = pd.read_csv("Data\JC-202010-citibike-tripdata.csv")
november_df = pd.read_csv("Data\JC-202011-citibike-tripdata.csv")
december_df = pd.read_csv("Data\JC-202012-citibike-tripdata.csv")


```

```python
# create a new dataframe and append all the dataframes
citi_2020_df = pd.DataFrame()

# # append all the dataframes
citi_2020_df = pd.concat([january_df, february_df, march_df, april_df, may_df, june_df, july_df, august_df, september_df, october_df, november_df, december_df])

citi_2020_df.info()

# # save the new dataframe to a csv file
# citi_2020_df.to_csv("Data\citi_2020_df.csv")


```

```python
# insert a column with the durartion as minutes
citi_2020_df['duration_minutes'] = round(citi_2020_df['tripduration'] / 60, 1)

citi_2020_df.head()



```
